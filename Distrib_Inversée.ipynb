```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Définition de la fonction de densité de probabilité (non normalisée)
def unormalized_pdf(x, sigma):
    return np.exp(-1 / (1 + x**2 / (2 * sigma**2)))  # Limitation de la croissance

# Paramètre de la distribution
sigma = 1.5  # Échelle de la distribution
L = 5  # Limite d'intégration pour éviter l'infini

# Calcul de la constante de normalisation
norm_const, _ = quad(lambda x: unormalized_pdf(x, sigma), -L, L)

# Normalisation de la fonction de densité
def normalized_pdf(x, sigma, norm_const):
    return unormalized_pdf(x, sigma) / norm_const

# Génération de points pour tracer la distribution
x_vals = np.linspace(-L, L, 1000)
y_vals = normalized_pdf(x_vals, sigma, norm_const)

# Tracé de la distribution
plt.figure(figsize=(8, 5))
plt.plot(x_vals, y_vals, label="Distribution Inversée", color='red')
plt.xlabel("x")
plt.ylabel("Densité de Probabilité")
plt.title("Distribution Opposée à la Normale (Extrêmes Fréquents)")
plt.legend()
plt.grid()
plt.show()
```
